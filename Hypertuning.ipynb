```python
import numpy as np
#kernels,gamma,C,cv
kernels=['linear','rbf']
C=[1,10,20]
avg_scores={}
for kval in kernels:
  for cval in C:
    cv_scores=cross_val_score(SVC(kernel=kval,C=cval,gamma='auto'),input,target,cv=5)
    #avg_scores[kval + '_' + str(cval)]= np.average(cv_scores)
    avg_scores[kval + '_' + str(cval)]=np.average(cv_scores)
avg_scores
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
model_params={
    'svm':{
        'model':SVC(gamma='auto'),
        'params':{
        'kernel':['linear','rbf'],
        'C':[1,10,20],
    }
    },
    'random_forest':{
        'model':
         RandomForestClassifier(),
    'params':{
        'n_estimators':[1,5,10]
    }
    },
    'logistic_regression':
    {
        'model':LogisticRegression(multi_class='auto',solver='liblinear'),
        'params':{
            'C':[1,5,10]
        }
    },

}
```

```python
scores=[]
for model_name,model_param in model_params.items():
  clf2=GridSearchCV(model_param['model'],model_param['params'],return_train_score=False)
  clf2.fit(input,target)
  scores.append({
      'model': model_name,
      'best_score': clf2.best_score_,
      'best params': clf2.best_params_
  })
```

```python
pd.DataFrame(scores)
```
